## Purchasing Analysis (Gender)

```python
only_male = file_pd.loc[(file_pd['Gender']=='Male')]
male_purchase_value = only_male['Price'].sum()
avg_male_purchase_price = male_purchase_value/count['Male']
only_female = file_pd.loc[(file_pd['Gender']=='Female')]
female_purchase_value = only_female['Price'].sum()
avg_female_purchase_price = female_purchase_value/count['Female']
only_other = file_pd.loc[(file_pd['Gender']=='Other / Non-Disclosed')]
other_purchase_value = only_other['Price'].sum()
avg_other_purchase_price = other_purchase_value/count['Other / Non-Disclosed']
# print(avg_male_purchase_price)
purchase_analysis_dict = {'Gender':['Male','Female','Other / Non-Disclosed'], 'Purchase Count':count,
                          'Average Purchase Price':[avg_male_purchase_price,avg_female_purchase_price,avg_other_purchase_price],
                          'Total Purchase Cost':[male_purchase_value,female_purchase_value,other_purchase_value]}

purchase_analysis_pd = pd.DataFrame(purchase_analysis_dict)
purchase_analysis_pd['Average Purchase Price'] = purchase_analysis_pd['Average Purchase Price'].map('${:,.2f}'.format)
purchase_analysis_pd['Total Purchase Cost'] = purchase_analysis_pd['Total Purchase Cost'].map('${:,.2f}'.format)
purchase_analysis_pd.set_index('Gender')
```

## Top Spenders

```python
spenders_gb = file_pd.groupby(by = 'SN')
purchase_count_spender = spenders_gb['Price'].count()
avg_purchase_cost_spender = spenders_gb['Price'].mean().map('${:,.2f}'.format)
total_purchase_price_spender = spenders_gb['Price'].sum().map('${:,.2f}'.format)
top_spenders_dict = {"Purchase Count": purchase_count_spender,"Average Purchase Price":avg_purchase_cost_spender,"Total Purchase Value":total_purchase_price_spender}
top_spenders_df = pd.DataFrame(data = top_spenders_dict, columns = ['Purchase Count','Average Purchase Price','Total Purchase Value'])
top_spenders_df.sort_values(by='Total Purchase Value', ascending= False).head()
```

##  Most Popular Items

```python
popular_gb = file_pd.groupby(by = ['Item ID','Item Name'])
purchase_count_popular = popular_gb['Price'].count()
avg_purchase_cost_popular = popular_gb['Price'].mean().map('${:,.2f}'.format)
total_purchase_price_popular = popular_gb['Price'].sum().map('${:,.2f}'.format)
popular_dict = {"Purchase Count": purchase_count_popular,"Average Purchase Price":avg_purchase_cost_popular,"Total Purchase Value":total_purchase_price_popular}
most_popular_df = pd.DataFrame(data = popular_dict, columns = ['Purchase Count','Average Purchase Price','Total Purchase Value'])
most_popular_df.sort_values(by='Purchase Count', ascending= False).head()
```
